# 建立TABLE

```python
import pymysql

host='127.0.0.1'
port=3306
user='root'
passwd='cvfgty12'
db = 'tapmcdatabase'
conn = pymysql.connect(host=host, port=port, user=user, passwd=passwd, db=db, charset='utf8')

cursor = conn.cursor()
#刪除TABLE
sql_drop_table = """
DROP TABLE IF EXISTS dailytrans;
"""
cursor.execute(sql_drop_table)
conn.commit()

#新增TABLE
sql_create_table = """
CREATE TABLE IF NOT EXISTS dailytrans(
    date DATE,
    market VARCHAR(45),
    supplier_code VARCHAR(45) NOT NULL,
    code VARCHAR(45),
    fruit VARCHAR(45),
    specification VARCHAR(45),
    grade VARCHAR(45),
    counts INT,
    net_weight DECIMAL(10, 2),
    unit_price INT,
    class VARCHAR(45),
    basket VARCHAR(45),
    ticket_num VARCHAR(45),
    ticket_cate VARCHAR(45),
    buyer VARCHAR(45) NOT NULL,
    salesperson VARCHAR(45),
    item VARCHAR(45)
);
"""
cursor.execute(sql_create_table)
conn.commit()

cursor.close()
conn.close()
```

# 插入資料

```python
import pandas as pd

def read_Sheet(file):
    path = r"D:/taipeifruit/taipeifruit/"
    excel_file = pd.ExcelFile(path+file)
    #讀取EXCEL檔，合併每個Sheet
    sheet_names = excel_file.sheet_names
    dfs_list = []
    for sheet in sheet_names:
        if sheet == sheet_names[0]:
            df = pd.read_excel(path+file, sheet_name=sheet, header=None, dtype={1:str,6:str,12:str,13:str,14:str,15:str,16:str})
            #刪除前三列
            df.drop([0,1,2,3], axis=0, inplace=True)
            df = df.reset_index(drop=True)
            dfs_list.append(df)
        elif sheet == sheet_names[-1]:
            df = pd.read_excel(path+file, sheet_name=sheet, header=None, dtype={1:str,6:str,12:str,13:str,14:str,15:str,16:str})
            #刪除後二列
            df.drop(df.index[-2:], axis=0, inplace=True)
            df = df.reset_index(drop=True) 
            dfs_list.append(df)
        else:
            df = pd.read_excel(path+file, sheet_name=sheet, header=None, dtype={1:str,6:str,12:str,13:str,14:str,15:str,16:str})
            dfs_list.append(df)            
    df = pd.concat(dfs_list, axis=0)
    df.reset_index(drop=True, inplace=True)
    return df

def Sheet_concat(df1, df2):
    df = pd.concat([df1, df2], axis=0)
    df.reset_index(drop=True, inplace=True)
    return df
```

# table取資料並存成Dataframe

```python
import pymysql
import pandas as pd

# 連線MySQL
host = '127.0.0.1'
port = 3306
user = 'root'
passwd = 'cvfgty12'
db = 'tapmcdatabase'

conn = pymysql.connect(host=host, port=port, user=user, passwd=passwd, db=db, charset='utf8')
cursor = conn.cursor()

# 執行SQL查詢
sql_query = "SELECT * FROM dailytrans;"
cursor.execute(sql_query)

# 獲取查詢結果
result = cursor.fetchall()

# 將結果轉換為DataFrame
df = pd.DataFrame(result, columns=[i[0] for i in cursor.description])

# 關閉連線
cursor.close()
conn.close()
```
