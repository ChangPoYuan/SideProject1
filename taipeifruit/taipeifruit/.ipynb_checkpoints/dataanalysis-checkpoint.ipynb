```python
#以日期、品種分群 811白肉，812紅肉；並計算分別的淨重
group = df.groupby(['dt', 'ft'])
nw_sum = group.agg({'nw':sum})
print(nw_sum.head(5))
print(nw_sum.index)
```

```python
index = pd.MultiIndex.from_tuples(nw_sum.index)
nw_sum = nw_sum.reindex(index)
nw_sum = nw_sum.unstack()
nw_sum
```

```python
nw_sum[nw_sum.isnull().any(axis=1)]
```

```python
#填補空值，NAN代表當日沒有該品項，所以銷售量為0

nw_sum = nw_sum.fillna(0)
nw_sum[nw_sum.isnull().any(axis=1)]
```

```python
#以月為單位進行統計 x代表品項, y代表月份, z代表要加總的column, w代表年份, arg為要迭代的dataframe

def do_monthly_sum(x,y,z,w,*args):
    list_1=[]
    list_2=[]
    list_3=[]
    list_4=[]
    list_5=[]
    list_6=[]
    list_7=[]
    list_8=[]
    list_9=[]
    list_10=[]
    list_11=[]
    list_12=[]
    for index, row in args[0].iterrows():
        index = index.replace('-', '')
        if int(w+'0100') < int(index) < int(w+'0200'):
            list_1.append(row[(str(z),str(x))])
        elif int(w+'0200') < int(index) < int(w+'0300'):
            list_2.append(row[(str(z),str(x))])
        elif int(w+'0300') < int(index) < int(w+'0400'):
            list_3.append(row[(str(z),str(x))])
        elif int(w+'0400') < int(index) < int(w+'0500'):
            list_4.append(row[(str(z),str(x))])
        elif int(w+'0500') < int(index) < int(w+'0600'):
            list_5.append(row[(str(z),str(x))])
        elif int(w+'0600') < int(index) < int(w+'0700'):
            list_6.append(row[(str(z),str(x))])
        elif int(w+'0700') < int(index) < int(w+'0800'):
            list_7.append(row[(str(z),str(x))])
        elif int(w+'0800') < int(index) < int(w+'0900'):
            list_8.append(row[(str(z),str(x))])
        elif int(w+'0900') < int(index) < int(w+'1000'):
            list_9.append(row[(str(z),str(x))])
        elif int(w+'1000') < int(index) < int(w+'1100'):
            list_10.append(row[(str(z),str(x))])
        elif int(w+'1100') < int(index) < int(w+'1200'):
            list_11.append(row[(str(z),str(x))])
        elif int(w+'1200') < int(index) < int(w+'1300'):
            list_12.append(row[(str(z),str(x))])
    list_13=[sum(list_1),sum(list_2),sum(list_3),sum(list_4),sum(list_5),sum(list_6),sum(list_7),sum(list_8),sum(list_9),sum(list_10),sum(list_11),sum(list_12)]
    for i in range(12):
        if y==i+1:
            return list_13[i]
```

```python
#811 110年12個月分別的銷售總重(單位公斤)
month_811_weight_110 = []
for i in range(12):
    month_811_weight_110.append(do_monthly_sum(811,i+1,'nw','110',nw_sum))

#812 110年12個月分別的銷售總重(單位公斤)
month_812_weight_110 = []
for i in range(12):
    month_812_weight_110.append(do_monthly_sum(812,i+1,'nw','110',nw_sum))
    
print(month_811_weight_110)
print(month_812_weight_110)
```

```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.font_manager import FontProperties

# 設定中文字體
font = FontProperties(fname='NotoSansTC-VariableFont_wght.ttf', size=12)
month = ['M1','M2','M3','M4','M5','M6','M7','M8','M9','M10','M11','M12']

def plot_bar(a,b,c,d):
    np1 = np.array(a)/1000
    np2 = np.array(b)/1000
    ind = np.arange(12)
    width = 0.25
    plt.bar(ind, np1, width, label='白龍果')
    plt.bar(ind+width, np2, width, label='紅龍果')

    plt.xlabel('月份', fontproperties=font)
    plt.ylabel(c, fontproperties=font)
    plt.title(d, fontproperties=font)
    plt.xticks(ind+width/2, month, rotation=45)
    plt.grid(True)
    plt.legend(prop=font, loc='best')
    plt.show()
    
def plot_time(a,b,c,d):
    np1 = np.array(a)/1000
    np2 = np.array(b)/1000
    plt.plot(month, np1, marker='o', linestyle='-', label='白龍果')
    plt.plot(month, np2, marker='o', linestyle='-', label='紅龍果')
    plt.xlabel('月份', fontproperties=font)
    plt.ylabel(c, fontproperties=font)
    plt.title(d, fontproperties=font)
    plt.xticks(rotation=45)
    plt.grid(True)
    plt.legend(prop=font, loc='best')
    plt.show()
```

```python
import operator
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

price = []
demand = []
for i, j, k, w in zip(month_811_revenue_110,month_812_revenue_110, month_811_weight_110, month_812_weight_110):
    x = i+j
    y = k+w
    price.append(x/y)
    demand.append(y)
    
for i, j, k, w in zip(month_811_revenue_111,month_812_revenue_111, month_811_weight_111, month_812_weight_111):
    x = i+j
    y = k+w
    price.append(x/y)
    demand.append(y)

for i, j, k, w in zip(month_811_revenue_112,month_812_revenue_112, month_811_weight_112, month_812_weight_112):
    x = i+j
    y = k+w
    if y > 0:
        price.append(x/y)
        demand.append(y)
        
sort_axis = operator.itemgetter(0)
sorted_zip = sorted(zip(demand,price), key=sort_axis)
demand, price = zip(*sorted_zip)
plt.scatter(demand, price, color='b', marker='*')

#找出價格需求曲線
demand_np = np.array(demand)
price_np = np.array(price)
demand_np = demand_np[:, np.newaxis]
price_np = price_np[:, np.newaxis]
polynomial_features= PolynomialFeatures(degree=3)
x_poly = polynomial_features.fit_transform(demand_np)
model = LinearRegression()
model.fit(x_poly, price_np)
y_poly_pred = model.predict(x_poly)
rmse = mean_squared_error(price_np,y_poly_pred)
r2 = r2_score(price_np,y_poly_pred)
print('rmse is {}'.format(rmse))
print('R2 is {}'.format(r2))
sort_axis = operator.itemgetter(0)
sorted_zip = sorted(zip(demand_np,y_poly_pred), key=sort_axis)
demand_np, price_np = zip(*sorted_zip)
plt.plot(demand_np, price_np, color='m')

plt.title('Price-Demand-Curve')
plt.ylabel('Price')
plt.xlabel('Demand')
plt.grid(True)
plt.show()
```

# XGboost Regression

```python
#不需要對資料標準化(因為XGBOOST內建正則化) #因為峰態呈現右偏，所以使用rmsle作為評估指標
from sklearn.model_selection import train_test_split
import pandas  as pd
import numpy   as np
import xgboost as xgb

X = df.drop(['aver_rev'], axis=1)
y = df.get('aver_rev')

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=8)

#=========================================================================
# XGBoost regression: 
# Parameters: 
# n_estimators  "Number of gradient boosted trees. Equivalent to number 
#                of boosting rounds."
# learning_rate "Boosting learning rate (also known as “eta”)"
# max_depth     "Maximum depth of a tree. Increasing this value will make 
#                the model more complex and more likely to overfit." 
#=========================================================================
# regressor=xgb.XGBRegressor(eval_metric='rmsle', tree_method='gpu_hist')
regressor=xgb.XGBRegressor(eval_metric='rmsle')

#=========================================================================
# exhaustively search for the optimal hyperparameters
#=========================================================================
from sklearn.model_selection import GridSearchCV
# set up our search grid
param_grid = {"max_depth":    [4, 5, 6, 7],
              "n_estimators": [300, 400, 500, 600],
              "learning_rate": [0.045, 0.05, 0.055, 0.06]}

## put code here
# gs = GridSearchCV(regressor, param_grid = param_grid, cv=5, scoring='neg_mean_squared_log_error', n_jobs=-1)
gs = GridSearchCV(regressor, param_grid = param_grid, cv=5, scoring='neg_mean_squared_log_error')
gs.fit(X_train, y_train)
print("The best hyperparameters are ",gs.best_params_)
print("The best model score is",gs.best_score_)
```

```python
from sklearn.metrics import mean_squared_log_error, r2_score
# put code here
# regressor_op=xgb.XGBRegressor(n_estimators = 600, max_depth = 5, learning_rate = 0.03, eval_metric='rmsle', tree_method='gpu_hist')
regressor_op=xgb.XGBRegressor(n_estimators = 500, max_depth = 5, learning_rate = 0.05, eval_metric='rmsle')
regressor_op.fit(X_train, y_train)
pred_train = regressor_op.predict(X_train)

print("The score is %.5f" %np.sqrt(mean_squared_log_error(y_train, pred_train)))
print("The score is %.5f" %r2_score(y_train, pred_train))
```

# 找出火龍果跟甚麼水果是購買組合(目的找尋新客戶)

```python
#選取date buyer fruit 
#使用Apriori演算法
from mlxtend.frequent_patterns import association_rules
from mlxtend.frequent_patterns import apriori

new_df = df.groupby(['date', 'buyer','fruit'])['counts'].sum().unstack().reset_index().fillna(0).set_index('date')
new_df.head(10)
```

```python
#因為購物籃分析不考慮商品購買數量，所以將數量>0的值轉為1

basket_df = new_df.drop('buyer', axis=1)
basket_df = basket_df.applymap(lambda x:1 if int(x)>0 else 0)
basket_df.head(10)
```

```python
#找出頻繁商品項目集>0.2的商品集

frequent_itemsets_fr = apriori(basket_df, min_support=0.2, use_colnames=True)
frequent_itemsets_fr[frequent_itemsets_fr['itemsets'].apply(lambda x: str(812) in x)].sort_values('support', ascending=False)
```

```python
#尋找lift>1正相關且結果商品為812

a_rules = association_rules(frequent_itemsets_fr, metric='lift', min_threshold=1)
a_rules[(a_rules['consequents'] == {'812'}) | (a_rules['consequents'] == {'811'})].sort_values('lift', ascending=False)
```

```python
#分析高頻率購買紅龍果的承銷人屬於哪個購買價位

print(df[df['fruit'] == '812']['unit_price'].describe())
```

```python
new_df = df[df['fruit'] == '812'][df[df['fruit'] == '812']['buyer'].isin(buyer_812list)]
new_df = new_df.groupby(['buyer'])['unit_price'].mean()
high = []
mediun = []
low = []
def level(x):
    if int(x) >= 62:      
        return '高價位'
    elif 62 > int(x) > 30:
        return '中價位'
    else:
        return '低價位'
a = new_df.apply(level).sort_values(ascending=False)
for i, j in a.iteritems():
    if j == '高價位':
        high.append(i)
    elif j == '中價位':
        mediun.append(i)
    else:
        low.append(i)
print("高價位：{}\n中價位：{}\n低價位：{}".format(high, mediun, low))
```
